## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# -------------------------------
# 1. Create copies of datasets
# -------------------------------
# Keeping original data safe
titles_df = titles.copy()
credits_df = credits.copy()
````

````python
# -------------------------------
# 2. Handle missing values
# -------------------------------

# Fill missing age certification with 'Not Rated'
titles_df['age_certification'] = titles_df['age_certification'].fillna('Not Rated')

# Drop rows where IMDb score is missing (important for analysis)
titles_df = titles_df.dropna(subset=['imdb_score'])

# Fill missing IMDb votes with 0
titles_df['imdb_votes'] = titles_df['imdb_votes'].fillna(0)
````

````python
# -------------------------------
# 3. Remove duplicate rows
# -------------------------------
titles_df = titles_df.drop_duplicates()
credits_df = credits_df.drop_duplicates()
````

````python
# -------------------------------
# 4. Clean genres column
# -------------------------------
titles_df['genres'] = titles_df['genres'].str.replace("[", "", regex=False)\
                                         .str.replace("]", "", regex=False)\
                                         .str.replace("'", "", regex=False)
````

````python
# -------------------------------
# 6. Ensure correct data types
# -------------------------------
titles_df['release_year'] = titles_df['release_year'].astype(int)
titles_df['runtime'] = titles_df['runtime'].astype(int)
titles_df['imdb_votes'] = titles_df['imdb_votes'].astype(int)
````

````python
# -------------------------------
# 7. Final check after data wrangling
# -------------------------------
titles_df.info()
titles_df.head()
````

#### Chart - 14 - Correlation Heatmap

````python
# Chart 14: Correlation Heatmap

# Select numerical columns for correlation
numerical_cols = titles_df[['release_year', 'runtime', 'imdb_score', 'imdb_votes']]

# Compute correlation matrix
corr_matrix = numerical_cols.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

# Add title
plt.title("Correlation Heatmap of Numerical Variables")

# Show plot
plt.show()
````

##### 1. Why did you pick the specific chart?

I selected a correlation heatmap because it is the most effective way to analyze relationships between multiple numerical variables at the same time. Instead of checking pairwise relationships one by one, a heatmap provides a single, visual summary showing the strength and direction of correlations.

##### 2. What is/are the insight(s) found from the chart?

The following insights were identified:

* **IMDb votes and IMDb score** show a moderate positive correlation, meaning titles that are more popular tend to receive slightly higher ratings.

* **Runtime has a very weak** or near-zero correlation with IMDb score, confirming that longer content does not necessarily result in better ratings.

* **Release year** shows minimal correlation with IMDb score, indicating that both older and newer content can perform equally well in terms of audience ratings.

* Overall, **content quality and audience engagement** influence ratings more than duration or release timing.
